#### KDE Plot: When thresh=0, the shading will extend to the very edges of where the KDE estimates any density, no matter how small. This results in the entire "blob" of the KDE being filled, making it look like it's shaded everywhere that the density estimation function produces a non-zero value.

#### KDE Plot: When thresh is a positive value, the shading will only begin from that specific density level upwards. Areas with a density lower than this threshold will remain unshaded (transparent). This is usefull for Ignoring "noise", focusing on core trends

#### Low thresh (closer to 0): You're seeing more of the distribution, including the very sparse, outer regions. Like a low water level, it covers more ground.

#### Higher thresh: You're focusing on the denser, more concentrated areas of the distribution, ignoring the "outskirts." Like a higher water level, it only fills the deeper basins.

### Facet grid
A FacetGrid is a powerful way to visualize the distribution of one or more variables across the levels of one or more categorical variables. It allows you to create a grid of plots where each plot (or "facet") shows a subset of your data based on the unique values of a categorical variable.

#### Why Use a FacetGrid?

Comparison: It makes it incredibly easy to compare patterns, distributions, or relationships across different subsets of your data. Instead of creating many individual plots manually, a FacetGrid automates the process.

Structure: It provides a highly structured and organized way to present multiple related plots.

Exploratory Data Analysis (EDA): It's a fundamental tool for EDA, allowing you to quickly uncover trends or differences that might be hidden when looking at aggregated data.

Consistency: All plots in the grid share the same axis limits (by default), making comparisons more reliable.

#### Common Scenarios for FacetGrid:

Trends over time for different categories: (e.g., stock prices for different companies over time)

Distribution of a variable across groups: (e.g., age distribution for different demographics)

Relationship between two variables across groups: (e.g., correlation between height and weight for different age groups)

### kws = dict(s=50, linewidth=0.5, edgecolor='black')
* Python dictionary named kws (short for "keyword arguments").
* s-> controls the size of the scatter points. So, each point will have a size of 50.
* linewidth=0.5: This controls the width of the outline around each scatter point.
* edgecolor='black': This sets the color of the outline around each scatter point to black.
* Purpose: This dictionary holds additional keyword arguments that you want to pass to the plotting function that g.map() will use. It's a clean way to organize multiple arguments.
* **kws: This is the "unpacking operator". It takes all the key-value pairs from your kws dictionary and passes them as individual keyword arguments to the plt.scatter function. So, plt.scatter will receive s=50, linewidth=0.5, and edgecolor='black'.

```python
g =sns.FacetGrid (movies, row = 'Genre', col = 'Year', hue = 'Genre')
kws = dict(s=50, linewidth=0.05,edgecolor='black')
g = g.map(plt.scatter, 'CriticRating', 'AudienceRating',**kws )
plt.show()
```
