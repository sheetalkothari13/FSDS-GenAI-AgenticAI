## use of  : 
- word_tokenize => Splits a sentence/text into words.
- sent_tokenize => Splits text into sentences.
- blankline_tokenize => Splits text into paragraphs (based on blank lines).
- WhitespaceTokenizer => Splits tokens only on spaces and tabs (does not separate punctuation)
- wordpunct_tokenize => Splits words and punctuation separately.
- from nltk.util import bigrams,trigrams,ngrams => sequence of n words/tokens.
- Stemming = reducing a word to its root form (may not be a valid word).
- porterStemmer => Classic and widely used, less aggressive.
- lancasterStemmer => More aggressive (cuts words shorter).
- SnowballStemmer => Improved and multilingual (supports 15+ languages).
- lemmatization => Reduces word to its dictionary form (lemma) using vocabulary + grammar. More accurate than stemming.
- NLU (Natural Language Understanding)=> Machine understands human language (intent, meaning). Used in chatbots, sentiment analysis, speech-to-text.
- NLG (Natural Language Generation)=> Machine produces human-like language (text, response). Used in text summarization, AI writing, report generation.

```python
from nltk import ne_chunk
```

```python
NE_sent = 'The US president stays in whitehouse'
```

```python
NE_tokens=word_tokenize(NE_sent)
NE_tokens
```

```python
NE_tags = nltk.pos_tag(NE_tokens)
NE_tags
```

```python
NE_ner = ne_chunk(NE_tags)
print(NE_ner)
NE_ner.draw()
```
