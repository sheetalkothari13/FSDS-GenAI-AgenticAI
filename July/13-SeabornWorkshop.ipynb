```python
sns.scatterplot(data = tips, x='total_bill',y='tip',hue = 'time',size='size',palette="deep")
plt.title("Scatter plot")
plt.show()
```

```python
sns.lineplot(data=tips,
             x = 'size', 
             y = 'total_bill',
            hue = 'sex',
            ci = None,
            palette='dark')
plt.title("line plot")
plt.show()
```

```python
sns.barplot(data = tips,
           x = 'day',
           y = 'total_bill',
           hue = 'sex',
           palette='muted')
plt.title("bar plot")
plt.show()
```

```python
sns.boxplot(data = tips,
            x = 'day',
            y = 'tip',
            hue = 'smoker',
            palette='pastel')
            
plt.title("boxplot")
plt.show()
```

```python
sns.violinplot(data = tips,
              x = 'day',
              y = 'total_bill',
              hue = 'time',
              split=True,
              palette='colorblind')
plt.title("Violin plot")
plt.show()
```

```python
sns.countplot(data = tips,
             x = 'day',
             hue = 'smoker',
             palette='bright')
plt.title('Count plot')
plt.show()
```

```python
sns.regplot(data = tips,
           x = 'total_bill',
           y = 'tip',
           scatter_kws={'s':50,'color':'green'},
           line_kws={'color':'red'})
plt.title('Regression plot')
plt.show()
```

```python
sns.histplot(data = tips,
            x = 'total_bill',
            bins = 20,
            kde = True,
            color = 'maroon')
plt.title("Histogram plot")
plt.show()
```

```python
sns.pairplot(tips,hue='sex',
            vars=["total_bill","tip","size"],
            palette='husl')
plt.suptitle("pair plot: Numerical variables by gender",y = 1.02)
plt.show()
```

```python
sns.catplot(data = tips,
            x = 'day',
            y = 'tip',
            hue = 'sex',
            kind = 'point',
            palette='bright'
)
plt.title('cat plot')
plt.show()


# kind : str
# The kind of plot to draw, corresponds to the name of a categorical
# axes-level plotting function. Options are: "strip", "swarm", "box", "violin",
# "boxen", "point", "bar", or "count".
```

```python
sns.catplot(data = tips,
            x = 'day',
            y = 'tip',
            hue = 'sex',
            kind = 'strip',
            palette='bright'
)
plt.title('cat plot')
plt.show()
```

```python
sns.catplot(data = tips,
            x = 'day',
            y = 'tip',
            hue = 'sex',
            kind = 'swarm',
            palette='bright'
)
plt.title('cat plot')
plt.show()
```

```python
sns.catplot(data = tips,
            x = 'day',
            y = 'tip',
            hue = 'sex',
            kind = 'boxen',
            palette='bright'
)
plt.title('cat plot')
plt.show()
```

```python
sns.jointplot(
    data = tips,
    x = 'total_bill',
    y = 'tip',
    kind = 'scatter',
    hue = 'sex',
    color = 'purple',
    palette = 'coolwarm'
)
plt.suptitle("joint plot",y = 1.02)
plt.show()
```

```python
 g = sns.FacetGrid(
        tips,
        col='time',
        row='smoker',
        margin_titles=True).map(sns.histplot,
                                 'total_bill',
                                 bins = 20,
                                 kde = True).add_legend()

plt.show()
```

```python
sns.stripplot(data=tips,
             x='day',
             y = 'tip',
             hue = 'sex',
             jitter=True,
             palette='Set1')
plt.title('stripplot')
plt.show()
```

```python
sns.kdeplot(data = tips,
           x = 'total_bill',
           hue = 'sex',
           # fill = True,
           palette='tab10')
plt.title('KDE plot')
plt.show()
```

```python
sns.kdeplot(data = tips,
           x = 'total_bill',
           hue = 'sex',
           fill = True,
           palette='tab10')
plt.title('KDE plot')
plt.show()
```
