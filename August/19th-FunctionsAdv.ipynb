# Passing list to function

```python
def count(lst):
    
    even = 0
    odd = 0
    
    for i in lst:
        if i%2 == 0:
            even += 1
        else:
            odd +=1 
    return even,odd

lst = [1, 2, 3, 4, 5 ]
even, odd = count(lst)         

print(even)
print(odd)
```

# Fibonacci Series

```python
def fib(n):
    a = 0
    b = 1
    
    print(a)
    print(b)
    
    for i in range(0, n):
        c = a + b
        a = b
        b = c
        
        print(c)
        
fib(5)
```

# Factorial

```python
def fact(n):
    f = 1
    for i in range(1, n+1):
        f = f*i
        
    return f

x = 4
result = fact(x)
print(result)  
```

```python
def fact(n):
    if n==0:
        return 1
    return n*fact(n-1)

result = fact(5)
result
```

```python
def fibb(n):
    if n<=1:
        return n
    else:
        return fibb(n-1)+fibb(n-2)

result=fibb(6)
result
```
