```python
plt.subplot(2, 1, 1)  #Divides figure into grid of subplots.Each subplot has its own axes.
plt.plot(x1, np.cos(x1), '*')  # first subplot of a 2-row grid.
plt.show()
#splits the figure into a grid of rows and columns, and lets you choose where to plot
#plt.subplot(nrows, ncols, index)
```

```python
plt.figure() #Create a new figure, or activate an existing figure.
plt.subplot(2,1,1)
plt.plot(x1, np.sin(x1))

plt.subplot(2, 1, 2)
plt.plot(x1, np.cos(x1))
plt.show()

plt.subplot(2, 1, 2)   #matplotlib automatically created a new figure (Figure B) in the background since Figure A was closed after plt.show().
plt.plot(x1, np.tan(x1))    #make a 2×1 grid in Figure B, and activate panel 2.
plt.show()

#Each subplot has its own axes, but they share the same x-range (0 to 10) 
#because they use the same x1 array.
# subplot() - want to organize multiple plots neatly in one figure instead of overlapping them
#Without subplot(), the second plot would overwrite the first.
```

```python
plt.figure() 
plt.subplot(2,1,1)
plt.plot(x1, np.sin(x1))

plt.subplot(2, 1, 2)
plt.plot(x1, np.cos(x1))
# plt.show()

plt.subplot(2, 1, 2)   # the cos and tan over witten on each other
plt.plot(x1, np.tan(x1))   
plt.show()
```

```python
plt.figure() 
plt.subplot(3,1,1)      # 3 rows mai divide kar, 1 column(axis), access its 1st panel(part)
plt.plot(x1, np.sin(x1))

plt.subplot(3, 1, 2)    #access is 2nd panel
plt.plot(x1, np.cos(x1))
# plt.show()

plt.subplot(3, 1, 3)   #access its 3rd panel
plt.plot(x1, np.tan(x1))   
plt.show()
```

```python
#Figure - is the top level container for all the plot elements. 
#We can think of the Figure object as a box-like container containing one or more Axes.
#Axes - represent an individual plot. 
#The Axes object contain smaller objects such as axis, tick marks, lines, legends, title and text-boxes.
```

```python
fig, ax = plt.subplots(2) #Creates a figure (fig) and an array of 2 axes objects (ax).
ax[0].plot(x1, np.sin(x1), 'b-')
ax[1].plot(x1, np.cos(x1), 'b-')
        #fig → the entire figure
        #ax  → array of Axes objects
        #ax[0] = Top subplot
        #ax[1] = Bottom subplot
plt.show()
```

```python
fig = plt.figure()

x2 = np.linspace(0, 5, 10)
y2 = x2 ** 2

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x2,y2, 'r')

axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title')

plt.show()

#fig = plt.figure().Creates a new blank figure object called fig.
#At this point, the figure has no axes (no plot area yet).

#Adds a set of axes (plot area) to the figure.
        #[left, bottom, width, height]
            #left = 0.1 → 10% from left edge of figure
            #bottom = 0.1 → 10% from bottom edge
            #width = 0.8 → 80% of figure width
            #height = 0.8 → 80% of figure height
    # creates a plot area that is centered inside the figure with a small margin.
```

```python
plt.style.use('ggplot')  # Apply ggplot theme

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave - ggplot style")
plt.show()
```

```python
plt.style.use(['dark_background', 'seaborn'])
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave - ggplot style")
plt.show()
```

```python
plt.style.use(['dark_background'])

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='sin(x)')
plt.title("Sine Wave with Combined Styles")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```
